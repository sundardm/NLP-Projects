```python
import json
import nltk

from collections import Counter
from prettytable import PrettyTable
import operator
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
import string
```

```python
with open('feed.json') as f:
    d = json.load(f)
```

```python
cont = []
for i in d.keys():
    cont.append(d[i]['content'])
    
```

```python
email_at = [] #emails at in our data
for i in cont:
    email_at.append((re.findall(r"(\@+[\w_]+)",i)))
email_at = [x for x in email_at if x != []]
email_at = [at.lower() for x in email_at for at in x]
email_at
```

```python
hash1 = [] #hashtags in our data
for i in cont:
    hash1.append((re.findall(r"(\#+[\w_]+)",i)))     #find all hashtags

hash1 = [x for x in hash1 if x != []]
hash1 = [hash.lower() for x in hash1 for hash in x]
hash1 
```

```python
#Removing stopwords and other characters as defined by me

words = []
sent = []
unwords = []
words1 = []
words10= []



stopw = stopwords.words('english') + list(string.punctuation)  +["’","”","—","“", "themasters", "april"] + email_at + hash1
stopw1 = stopwords.words('english') +["’","”","—","“", "themasters", "april"] + email_at + hash1

for i in range(0,len(cont)):
    count_all = Counter()
    x = cont
    text = re.sub(r'^https?:\/\/.*[\r\n]*', '', x[i])
   
    #Words
    tokens = [stop for stop in word_tokenize(text.lower()) if stop not in stopw] #removes stopwords, punctuation and custom filter
    tokens = [w for w in tokens if w.isalpha()]  #remove every that is not fully alphabetic
    tokens = [w for w in tokens if re.findall(r"(\w{3})",w) ] #removes anything less than 3 characters
    print("The number of words in entry number", i+1, " is", len(tokens))
    
    #Sentence
   
    sents = sent_tokenize(text)
    sents_rm_stopwords = []
    for sent in sents:
        tokens1 = [stop for stop in word_tokenize(sent.lower()) if stop not in stopw1] #removes stopwords, punctuation and custom filter
        tokens1 = [w for w in tokens1 if w.isalpha()]  #remove every that is not fully alphabetic
        tokens1 = [w for w in tokens1 if re.findall(r"(\w{3})",w) ] #removes anything less than 3 characters
        sents_rm_stopwords.append(' '.join(tokens1))
    print("The number of sentences in entry number", i+1, " is", len(sents_rm_stopwords))
    count_all.update(tokens)
    
    
    #unique words
    print("The number of unique words in entry number", i+1, " is", len(count_all))
 
    #Words appear once
    keybank = []
    for key,value in count_all.items():
        
        if value == 1:
            
            keybank.append(key)
    print("The number of words that appear only once in entry number", i+1, "is", len(keybank))
    
    
    #10 most common
    
    print("The 10 most commons words in entry number", i+1, " is", count_all.most_common(10))
```

```python
extractor = Extractor(extractor = 'LargestContentExtractor', url = URL)
```

```python
bank = sent_tokenize(extractor.getText())
```

```python
corpus = {'a': bank[1],
         'b': bank[2],
         'c': bank[3]}
```

```python
corpus
```

### TFIDF



```python

from math import log
import numpy as np

query = ['dartmouth', '250th']

def tf(query, corpus, normalize=True):
    
    TF = {}
    for j in query:
        tf = []
        for i in corpus:
            count = 0
            for k in corpus[i].lower().split(): 
                if j in k:
                    count += 1
            tf.append(count/len(corpus[i].lower().split())) 
            TF[j] = tf
    return TF
    
TF = tf(query, corpus)
TF
```

```python
def idf(query, corpus):
    IDF = {}
    for j in query:
        idf = []
        count = 0
        for i in corpus:
            if j in corpus[i].lower():
                count += 1
        idf.append(1 + log(len(corpus)/count))    
        IDF[j] = idf  
    return IDF

IDF = idf(query, corpus)
IDF
```

```python
def tf_idf(TF, IDF):
    TFIDF ={}
    for i in query:
        save = []
        x = np.multiply(TF[i], IDF[i])
        save.append(x)
        TFIDF[i] = save
    return TFIDF

TFIDF = tf_idf(TF, IDF)
TFIDF
```

```python
np.add(TFIDF['dartmouth'],TFIDF['250th'])
```

The combined vector TF-IDF scores for 'dartmouth' and '250th' among the 3 setences in our corpus is comparable where the scores for sentence 2 and 3 are very similar and the TF-IDF score assigns almost equal importance to both sentences within our corpus. (Sentence 2 is marginally more important). Sentence 1's combined score is highest among the 3 and as a result it has the highest importance. We are most likely to find a document that has either the query word 'dartmouth' or '250th' as being important in that sentence. What should be noted is that since we summed the TF-IDF scores for the indivudal query words, this vector output does not denote the combined importance of a term within that sentence, but instead a possibility where they both might be important, or just one of them might be important. This is evident when we look at sentence two, where 250th has no importance (no occurence) and has a TF-IDF score of 0. However, 'dartmouth' has our highest importance (occurence with a high % of occurence/total occurence) among all our sentences. This effect is reduced by the fact of '250th' having no occurences and hence places it 2nd overall on our summed TF-IDF rank. This is because in sentence 1, both 'dartmouth' and '250th' has a relatively good importance level, and hence their sum ranks 1. As a result, with our query words, sentence 1 has the highest rank and understandably so. 

## Scraping HTML Data

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
url="https://www.rottentomatoes.com/browse/box-office/"
response=requests.get(url)
```

```python
soup=BeautifulSoup(response.content,'html5lib')
```

```python
seven = soup.find_all(class_ = "panel-body content_body")[0]
```

```python
thisweek = []
lastweek = []
tmeter = []
title = []
weeks = []
gross =[]
totg = []
tavg = []
no = []

eight = seven.find_all(itemprop = "itemListElement")

for i in eight:
    nine = i.find_all("td")
    thisweek.append(nine[0].get_text().strip())
    lastweek.append(nine[1].get_text().strip())
    tmeter.append(nine[2].get_text().strip())
    title.append(nine[3].get_text().strip())
    weeks.append(nine[4].get_text().strip())
    gross.append(nine[5].get_text().strip())
    totg.append(nine[6].get_text().strip())
    tavg.append(nine[7].get_text().strip())
    no.append(nine[8].get_text().strip())
```

```python
pd.DataFrame({"This Week": thisweek, 
             "Last Week": lastweek,
             "T-Meter": tmeter,
             "Title": title,
             "Weeks Released": weeks,
             "Weekend Gross": gross,
             "Total Gross": totg,
             "Theatre Average": tavg,
             "# of Theatres": no})
```
